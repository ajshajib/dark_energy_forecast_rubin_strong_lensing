```python
import numpy as np
import matplotlib.pyplot as plt

from plot_util import *

set_fontscale(2)
```

```python
n_total = 120000
n_total_conservative = 65000

year_steps = np.array([1, 2, 5, 10])

n_total_years = n_total * np.sqrt(year_steps / 10)
n_total_conservative_years = n_total_conservative * np.sqrt(year_steps / 10)
```

```python
n_total_years / n_total_years[-1] * 100
```

```python
figure_width_inches = page_width_in * 0.75 * 2
# golden ratio
height_ratio = 1.61803398875

fig = plt.figure(figsize=(figure_width_inches, figure_width_inches / height_ratio))
ax = fig.add_subplot(111)

ax.bar(
    np.arange(len(year_steps)),
    n_total_years,
    color=cb2_orange,
    label="Optimal",
)
ax.bar(
    np.arange(len(year_steps)),
    n_total_conservative_years,
    label="Conservative",
    color=cb2_blue,
)
ax.set_xlabel("LSST Dataset")
ax.set_ylabel(r"$N$ of galaxy–galaxy lenses")
ax.set_xticks(np.arange(len(year_steps)))
ax.set_xticklabels([f"Year {y}" for y in year_steps])
# add comma to y-axis labels
ax.set_yticks([20000, 40000, 60000, 80000, 100000, 120000])
ax.get_yaxis().set_major_formatter(
    plt.FuncFormatter(lambda x, loc: "{:,}".format(int(x)))
)
ax.legend()

# fig.tight_layout()

fig.savefig("figures/fig1.pdf", bbox_inches="tight")
```
